# Matplotlib Tutorial

## Importing the required Modules


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Plotting a Simple Graph

```python
x=[1,2,3,4]
y=[10,3,2,4]
x2=np.arange(1,10,2)
plt.figure(figsize=(5,3),dpi=100)
plt.plot(x,y,'r.:',label='Line 1',linewidth=3,markersize=15,markeredgecolor="b",markeredgewidth=2)
plt.plot(x2,x2*2,'y*-.',label="Line 2",markerfacecolor="b",linewidth=4,markersize=7,markeredgecolor='r',markeredgewidth=3)
plt.title("First Graph",fontdict={'fontname':"Times New roman",'fontsize':"22"})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks([i for i in range(0,15,2)])
plt.yticks([i for i in range(0,25,4)])
plt.legend()
plt.savefig('firstpic.pdf')

plt.show()
```

## Real World Example

### Using the plot class to use the gas prices of the various year and country

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("gas_price.csv")
df

```

```python
plt.plot(df['Year'],df['Australia'],'yo--',label="Australia",linewidth=3)
plt.plot(df.Year,df.France,'c1:',label="France",linewidth=3,markeredgewidth=4)
plt.plot(df.Year,df.UK,'g+--',label="UK",linewidth=3)
plt.plot(df.Year,df.USA,'r*-.',label="USA",linewidth=3)
plt.title("Gas Prediction")
plt.xticks(df['Year'][::3].tolist())
plt.xlabel("Year")
plt.ylabel("Price in USD")
plt.legend(bbox_to_anchor=(1,1),title="Country")
plt.grid(linewidth=3)
plt.show()
```

## Pie Chart
simple pie chart

```python
values=np.random.randint(10,20,5)

plt.pie(values,labels=values,explode=[0.1,0,0,0,0.3],colors=['r','g','b','y','c'],radius=1,autopct='%d  %%')
plt.title("Simple Pie chart")
plt.legend()
plt.show()
```

## Using the Subplot Function

```python
spain=fifa.loc[fifa.Nationality=="Spain"].count()[0]

belgium=fifa.loc[fifa.Nationality=="Belgium"].count()[0]
sweden=fifa.loc[fifa.Nationality=="Sweden"].count()[0]
portugal=fifa.loc[fifa.Nationality=="Portugal"].count()[0]
plt.suptitle("Pie and Bar Chart")
plt.subplot(1,2,1)
plt.title("Players in country")
plt.pie([spain,belgium,sweden,portugal],labels=["Spain","Belgium","Sweden","Portugal"],autopct="%d %%",shadow=True,explode=[0,0,0.2,0])
plt.legend(title="Players From Country",bbox_to_anchor=(0,0))
plt.show()
plt.subplot(1,2,1)
colrs=['r','b','c','g']
i=0
bars=plt.bar(["Spain","Belgium","Sweden","Portugal"],[spain,belgium,sweden,portugal])
for ba in bars: 
    ba.set_color(colrs[i])
    i=i+1
plt.yticks(np.arange(0,1100,100),minor=True)
plt.legend()
plt.show()
```

### Annotating the Text 
Writing the Text and Mathematical Expression

```python
x=np.arange(30,87,4)
print(x.size)
y=np.arange(20,40,1.4)
print(y.size)
plt.scatter(x,y,marker=">",label="X vs Y")
plt.text(20,5,"Here is the text")  # simple text
plt.text(20,30,r"$\omega\ \pi^2$",fontdict={'fontname':'Cambria','fontsize':20})  # mathematical expression
plt.annotate("Annotation",xy=(40,15),arrowprops={'facecolor':"black","shrink":0.05},xytext=(2,5))
plt.scatter(df.Canada,df.USA,marker="*",label="C VS USA")
plt.legend()
plt.show()
```

**Exercise 1:** Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot. gridline style should .

(_Take the help of the documentation_)

The scatter plot should look like this:

```python
import pandas as pd
import matplotlib.pyplot as plt

company=pd.read_csv("company_sales.csv")
toothpaste=company.toothpaste
month=company.month_number
plt.scatter(month,toothpaste,label="Toothpaste Sales per Month")
plt.grid(linestyle="--")
plt.title("Toothpaste Sales per Month")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.xticks(month)
plt.legend()
plt.show()

```

**Exercise 2**: Read face cream and facewash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
# the modules and the csv data has been already readen
facec=company.facecream
facew=company.facewash
plt.bar(month,facec,label="FaceCream Sales",align="edge",width=0.25)
plt.bar(month,facew,label="Facewash  Sales",align='edge',width=-0.25)
plt.xticks(month)
plt.grid()
plt.legend()
plt.show()
```
